```python
import matplotlib.pyplot as plt
from torchvision import datasets
import torch
test_DS = datasets.MNIST(root="data/",train=False, download=True) #
```

```python
print(test_DS.data.shape) #10000 black and white photos that are 28x28 in size.
print(test_DS.data[0]) # First photo data
```

```python
plt.imshow(test_DS.data[0]) #image show
```

```python
plt.imshow(test_DS.data[0],cmap="gray") #image show in black and white
```

```python
plt.imshow(test_DS.data[4],cmap="gray")
plt.colorbar()
```

```python
for i in range(6):
    plt.subplot(2,3,i+1)
    plt.imshow(test_DS.data[i],cmap="gray")
```

```python
for i in range(6):
    plt.subplot(2,3,i+1, xticks=[],yticks=[]) # Delete axis
    plt.imshow(test_DS.data[i],cmap="gray")
```

```python
plt.figure(figsize=(7,4)) # Size change
for i in range(6):
    plt.subplot(2,3,i+1, xticks=[],yticks=[]) # Delete axis
    plt.imshow(test_DS.data[i],cmap="gray")
    plt.title(test_DS.targets[i], color="g")
```

```python
plt.figure(figsize=(7,4)) # Size change
for i in range(6):
    plt.subplot(2,3,i+1, xticks=[],yticks=[]) # Delete axis
    plt.imshow(test_DS.data[i],cmap="gray")
    plt.title(test_DS.targets[i].item(), color="g") # When using 'item', Delete 'Tensor' above each photos
```

```python
image=test_DS.data[0]
plt.figure(figsize=(10,10))
plt.imshow(image, cmap="gray")
plt.colorbar()
```

```python
image=test_DS.data[0]
plt.figure(figsize=(10,10))
plt.imshow(image, cmap="gray")
plt.colorbar()
print(image.shape)

for i in range(image.shape[0]):
    for j in range(image.shape[1]):
        plt.text(i,j,image[i,j].item(),horizontalalignment="center")
        # Display each values of pixels
```

```python
image=test_DS.data[0]
plt.figure(figsize=(10,10))
plt.imshow(image, cmap="gray")
plt.colorbar()
print(image.shape)

for i in range(image.shape[0]):
    for j in range(image.shape[1]):
        plt.text(i,j,image[i,j].item(),horizontalalignment="center",
                 color="k" if image[i,j] > torch.max(image)/1.5 else "w")
        # Display each values of pixels
```
