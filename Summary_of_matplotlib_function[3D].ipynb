```python
import matplotlib.pyplot as plt
import numpy as np
```

## 2 variables function, see as contour

```python
a=np.linspace(-1,1,100)
b=np.linspace(-1,1,100)
A, B=np.meshgrid(a,b) # display values of a and b in matrix
Loss=A**2 + B**2
plt.contour(a,b,Loss, 10) # 10 is the number of lines
plt.xlabel("input")
plt.ylabel("Loss") ; plt.grid()
```

## See in 3 dimensions

```python
plt.figure(figsize=[5,5]) # figsize can adjust size of figure
p = plt.axes(projection="3d") # this line can show us 3d coordinates
p.plot_surface(A,B,Loss) #thie line can draw graph in 3 dimensions

plt.figure(figsize=[7,7]) # figsize can adjust size of figure
p = plt.axes(projection="3d") # this line can show us 3d coordinates
p.plot_wireframe(A,B,Loss) # wireframe show us graph in different way
```

```python
plt.figure(figsize=[5,5]) # figsize can adjust size of figure
p = plt.axes(projection="3d") # this line can show us 3d coordinates
p.view_init(elev=10, azim=-10) # this line can adjust point of view
p.plot_surface(A,B,Loss) #thie line can draw graph in 3 dimensions

plt.figure(figsize=[7,7]) # figsize can adjust size of figure
p = plt.axes(projection="3d") # this line can show us 3d coordinates
p.view_init(elev=50, azim=-70) # this line can adjust point of view
p.plot_wireframe(A,B,Loss) # wireframe show us graph in different way
```

```python
from matplotlib.colors import LightSource # Color option is added
plt.figure(figsize=[10,9])
p = plt.axes(projection="3d") # this line can show us 3d coordinates
view= LightSource(altdeg=10, azdeg=-10) # this line can adjust point of view, deg means degree
color = view.shade(Loss, plt.cm.RdYlBu)
p.plot_surface(A,B,Loss, facecolors=color) #thie line can draw graph in 3 dimensions
```

```python
#from matplotlib.colors import LightSource # Color option is added
plt.figure(figsize=[10,9])
p = plt.axes(projection="3d") # this line can show us 3d coordinates
#view= LightSource(altdeg=10, azdeg=-10) # this line can adjust point of view, deg means degree
#color = view.shade(Loss, plt.cm.RdYlBu)
#p.plot_surface(A,B,Loss, facecolors=color) #thie line can draw graph in 3 dimensions
p.plot_surface(A,B,Loss,cmap='viridis')
```

```python
a=np.linspace(-1,1,10)
b=np.linspace(-1,1,10)
A, B=np.meshgrid(a,b) # display values of a and b in matrix
Loss=A**2 + B**2
plt.figure(figsize=[8,8])
p = plt.axes(projection="3d")
p.plot(A.reshape(-1,1).squeeze(),B.reshape(-1,1).squeeze(),Loss.reshape(-1,1).squeeze(),"ro")
#reshape change dimension of matrix and '-1' means that the number of row is set automatically by the number of column
#squeeze() delete dimension whose size 1
```

```python
a=np.linspace(-1,1,10)
b=np.linspace(-1,1,10)
A, B=np.meshgrid(a,b) # display values of a and b in matrix
Loss=A**2 + B**2
plt.figure(figsize=[8,8])
p = plt.axes(projection="3d")
p.scatter(A,B,Loss ,s=100, c='g',marker='*', edgecolor='b',alpha=0.9)
#scatter is simpler, s=size, c=color, alpha= gradation
```

```python
plt.figure(figsize=[8,6])
p=plt.axes(projection="3d")
p.plot_surface(A,B,Loss, alpha =0.5, cmap="viridis")
p.scatter(A,B,Loss,s=77,c="g",marker='o') # It's possible to show multiple graphs
```

#plotly

```python
import plotly.graph_objects as go

a=np.linspace(-3,3,50)
b=np.linspace(-3,3,50)
A,B = np.meshgrid(a,b)
print(A)
Cost = np.cos(A*1.5)+np.sin(B**2.4)
fig =go.Figure(data=[go.Surface(x=a, y=b, z=Cost, colorscale="viridis", opacity=0.75)])
fig.update_traces(contours_z=dict(show=True, usecolormap=True, highlightcolor="limegreen", project_z=True)) # Add Contour
fig.update_layout(title="test",width=500, height=500) # If I use this code, I can move view of point by using mouse
# When you hover over the graph, you can see the values
```

```python
fig =go.Figure(data=[go.Surface(x=a, y=b, z=Cost, colorscale="viridis", opacity=0.75)])
fig.update_traces(contours_z=dict(show=True, usecolormap=True, highlightcolor="limegreen", project_z=True)) # Add Contour
fig.update_layout(title="test",width=500, height=500, scene = dict(zaxis=dict(nticks=40, range=[-3,3])))
# The above code splits the -3 to 3 into 40 pieces
```
